# Skoog chapter 14, problem 43a

We can't ignore dilution for this one like we usually do.  The added volumes are significant compared to the intial volume.

I wrote out the equations in the scanned solutions

```octave
%plot -s 600,500 -f 'svg'

% before equiv point ------
Ka=3e-8; % Appendix 3 of text
Nastock=0.1; Vi=0.050; ATi=0.1;
vol=[0 0.005 0.015 0.0250 0.040 0.045 0.049];
Na=(Nastock*vol)./(Vi+vol);
AT=(ATi*Vi*ones(size(vol)))./(Vi+vol);
a=1*ones(size(vol)); b=Ka+Na; c=Na*Ka-Ka*AT;

for i=1:size(vol,2)
t=roots([a(i) b(i) c(i)]);
t=t(imag(t)==0);
t=t(t>0);
pH(i)=-log10(t);
end

```

```octave
%plot -s 600,500 -f 'svg'

% after equiv point ------
volafter=[0.05 0.051 0.055 0.06];
Na=(Nastock*volafter)./(Vi+volafter); 
Kb=(1e-14)/Ka;
AT=(ATi*Vi*ones(size(volafter)))./(Vi+volafter);
a=1*ones(size(volafter)); b=AT+Kb-Na; c=-Kb*Na;

for i=1:size(volafter,2)
t=roots([a(i) b(i) c(i)]);
t=t(imag(t)==0);
t=t(t>0);
pOH=-log10(t); pHafter(i)=14-pOH;
end

plot(volafter,pHafter,'ko','markersize',4,'markerfacecolor','b')
hold on
plot(vol,pH,'ko','markersize',4,'markerfacecolor','b')
```
